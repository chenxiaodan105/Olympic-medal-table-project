```python
#1992
#this part would show how I get data online and write it into a csv in detail for 1992
#read the medal table in 1992 from Wikipedia
medal_table1992=pd.read_html('https://en.wikipedia.org/wiki/1992_Summer_Olympics_medal_table')
medal_table1992=medal_table1992[1]
medal_table1992_2=medal_table1992.iloc[1:,1:]
```

```python
#change column names
medal_table1992_2.columns=['NOC','Gold1992','Silver1992','Bronze1992','Total1992']
medal_table1992_2.head()
```

```python
#split the country name into a country name and an abbreviation
split=DataFrame((x.split('\xa0') for x in medal_table1992_2['NOC']),
                index=medal_table1992_2.index,columns=['country','abbr'])
```

```python
medal_table1992_2=medal_table1992_2.iloc[:,1:]
medal1992=pd.concat([split,medal_table1992_2],axis=1)
medal1992.head()
```

```python
medal1992=medal1992.loc[:,['country','Gold1992','Silver1992','Bronze1992','Total1992']]
medal1992=medal1992.set_index('country')
```

```python
#choose 10 countries
countries=['United States','Great Britain','China','Russia','Germany',
           'Japan','France','South Korea','Italy','Australia']
```

```python
medal1992=medal1992.loc[countries,:]
medal1992
```

```python
#save the table as csv in the file
medal1992=medal1992.to_csv('medal1992.csv')
```

```python
#Qestion7: how many times for each country for hosting the Olympic games since 1900?
```

```python
#exploration 
host_nation=pd.read_html('https://en.wikipedia.org/wiki/List_of_Olympic_medals_by_host_nation')
```

```python
#get the target table
host=host_nation[0]
host.columns=['Year','Host Nation','Gold','Silver','Bronze','Total','Rank']
host=host.iloc[1:,:]
```

```python
split=DataFrame((x.split('\xa0') for x in host['Host Nation']),index=host.index,columns=['country','abbr'])
host2=pd.concat([split,host],axis=1)
```

```python
host=host.to_csv('host_nation.csv')
```

```python
plt.figure()
host3=host2.iloc[:28,0:5]
host3=host3.drop(['Host Nation','abbr','Year'],axis=1)
```

```python
#count how many times each country being the host nation by using groupby
host3=host3.groupby('country').count()
host3.columns=['Times']
host3
```

```python
host3.plot.bar(width=.5,figsize=(16,10),fontsize=14,color='navy')
plt.title('How many times for hosting the Olympic games by country since 1900')
plt.show()
```
